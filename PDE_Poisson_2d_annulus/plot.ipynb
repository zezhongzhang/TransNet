```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_pinn = pd.read_csv(f'pinn.csv')
print(df_pinn)

df_ls = pd.read_csv(f'ls.csv')
print(df_ls)
```

```python
plt.figure(figsize=(8,6))
plt.plot(df_ls['basis_num'], df_ls['mse_dnn_default'],'*-' ,label='dnn_default')
plt.plot(df_ls['basis_num'], df_ls['mse_pde_basis'],'*-' ,label='pde')

plt.plot(df_pinn['basis_num'], df_pinn['mse_final'],'*-', label='final_PINN')
plt.plot(df_pinn['basis_num'], df_pinn['mse_adam'],'*-', label='ADAM_PINN')
plt.yscale('log')
plt.legend()
plt.grid(which='both')
plt.xlabel('Basis_num')
plt.ylabel('Test MSE')
plt.title('Poisson 2d (annulus)')
plt.show()
```
