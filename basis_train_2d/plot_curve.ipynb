```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cor_len = 0.5
# cor_len = 1


r = 1.5
shape_vec = np.arange(0.2,5,0.2)
print(f'cor_{cor_len}_r_{r}')
df = pd.read_csv(f'raw/cor_{cor_len}_r_{r}_fine.csv')
# print(df)

n_rows = df.shape[0]
basis_num_vec = df['basis_num']
```

```python
plt.figure(figsize=(8,6))
for i in range(n_rows):
    mean_rmse = df.iloc[i,1:]
    loc = np.argmin(mean_rmse)
    plt.plot(shape_vec, mean_rmse, label=f'basis_num={basis_num_vec[i]}')
    plt.plot(shape_vec[loc], mean_rmse[loc],'r*')


plt.yscale('log')
plt.xlabel('shape')
plt.ylabel('RMSE')
plt.legend()
plt.grid(which='both')
plt.title('correlation length=0.5 (2d basis)')
plt.show()
```
